<h1> Préparer les données pour la modélisation </h1>

<h2> Label et OneHot Encoding </h2>

```python
labels_encode = ["frequence_deplacement"]
category_labels_encode = {"frequence_deplacement": ["Aucun", "Occasionnel", "Frequent"]}
onehot_labels = ["statut_marital", "departement", "poste", "domaine_etude"]


def encode_dataframe(
    df: pd.DataFrame,
    labels_encode: list,
    category_labels_encode: dict,
    onehot_labels: list,
    scale: bool = False,
    target_col: str = None  # colonne cible à exclure du scaling
) -> pd.DataFrame:
    """
    Encode les colonnes catégorielles en LabelEncoding (ordinal) ou OneHotEncoding,
    et optionnellement scale les colonnes numériques pour la modélisation,
    en excluant la colonne target.

    Args:
        df (pd.DataFrame): DataFrame d'entrée.
        labels_encode (list): Liste des colonnes à encoder avec LabelEncoding.
        category_labels_encode (dict): Dict {colonne: [liste ordonnée de catégories]}.
        onehot_labels (list): Liste des colonnes à encoder en OneHot.
        scale (bool): Si True, scale les colonnes numériques après encodage.
        target_col (str): Nom de la colonne cible à ne pas scaler.

    Returns:
        pd.DataFrame: DataFrame encodé (et éventuellement scale).
    """
    df_encoded = df.copy()

    # ---------------- LabelEncoding ----------------
    if labels_encode:
        ordinal_encoder = OrdinalEncoder(
            categories=[category_labels_encode[col] for col in labels_encode]
        )
        df_encoded[labels_encode] = ordinal_encoder.fit_transform(
            df_encoded[labels_encode]
        )

    # ---------------- OneHotEncoding ----------------
    if onehot_labels:
        ohe = OneHotEncoder(sparse_output=False, drop=None)
        ohe_encoded = ohe.fit_transform(df_encoded[onehot_labels])

        # récupérer les noms des nouvelles colonnes
        ohe_columns = ohe.get_feature_names_out(onehot_labels)
        df_ohe = pd.DataFrame(ohe_encoded, columns=ohe_columns, index=df.index)

        # supprimer les anciennes colonnes et concat
        df_encoded = df_encoded.drop(columns=onehot_labels)
        df_encoded = pd.concat([df_encoded, df_ohe], axis=1)

    # ---------------- Scaling optionnel ----------------
    if scale:
        numeric_cols = df_encoded.select_dtypes(include=['number']).columns.tolist()
        if target_col and target_col in numeric_cols:
            numeric_cols.remove(target_col)  # exclure la target du scaling
        scaler = StandardScaler()
        df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])

    return df_encoded

df_encoded = encode_dataframe(
    df_data_rh_join, labels_encode, category_labels_encode, onehot_labels, scale=True, target_col=target
)

#revenu mensuel très corrélé au niveau hiérarchie, supprimer niveau hoérarchie
df_encoded = df_encoded.drop(columns=["niveau_hierarchique_poste"])
```

<h1> Optimiser le modèle </h1>

<h2> Nouvelles features </h2>

```python
df_neaw_features = df_data_rh_join.copy()

# Performance
df_neaw_features['delta_note'] = df_neaw_features['note_evaluation_actuelle'] - df_neaw_features['note_evaluation_precedente']
# df_neaw_features['note_ponderee'] = (df_neaw_features['note_evaluation_actuelle'] * df_neaw_features['annees_dans_le_poste_actuel'] +
#                         df_neaw_features['note_evaluation_precedente'] * (df_neaw_features['annee_experience_totale'] - df_neaw_features['annees_dans_le_poste_actuel'])) / (df_neaw_features['annee_experience_totale'] + 1)

# Satisfaction
satisfaction_cols = ['satisfaction_employee_environnement', 
                        'satisfaction_employee_nature_travail', 
                        'satisfaction_employee_equipe', 
                        'satisfaction_employee_equilibre_pro_perso']
df_neaw_features['satisfaction_moyenne'] = df_neaw_features[satisfaction_cols].mean(axis=1)
df_neaw_features['satisfaction_std'] = df_neaw_features[satisfaction_cols].std(axis=1)

# Ancienneté
df_neaw_features['ratio_anciennete_poste'] = df_neaw_features['annees_dans_le_poste_actuel'] / (df_neaw_features['annee_experience_totale'] + 1)
df_neaw_features['ratio_temps_derniere_promo'] = df_neaw_features['annees_depuis_la_derniere_promotion'] / (df_neaw_features['annee_experience_totale'] + 1)

# Développement
df_neaw_features['engagement_developpement'] = df_neaw_features['nombre_participation_pee'] + df_neaw_features['nb_formations_suivies']
df_neaw_features['formations_par_annee'] = df_neaw_features['nb_formations_suivies'] / (df_neaw_features['annee_experience_totale'] + 1)

# Salaire / heures
df_neaw_features['revenu_par_heure_supp'] = df_neaw_features['revenu_mensuel'] / (df_neaw_features['heure_supplementaires'] + 1)
df_neaw_features['augmentation_relative'] = df_neaw_features['augementation_salaire_precedente'] / (df_neaw_features['revenu_mensuel'] + 1)


# Score global
df_neaw_features['score_global'] = df_neaw_features['satisfaction_moyenne'] + df_neaw_features['delta_note']

# ------------------ Supprimer les anciennes features redondantes ------------------
drop_cols = ['note_evaluation_precedente', 'note_evaluation_actuelle'] + satisfaction_cols + \
            ['annees_dans_le_poste_actuel', 'annee_experience_totale', 'nombre_participation_pee', 
                'nb_formations_suivies', 'revenu_mensuel', 'heure_supplementaires', 'augementation_salaire_precedente']

df_neaw_features.drop(columns=drop_cols, inplace=True)

df_encoded_new_features = encode_dataframe(
    df_neaw_features, labels_encode, category_labels_encode, onehot_labels, scale=True, target_col=target
)
```
